```python
# Import standard libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
sns.set_context('talk', font_scale = 1.05)
```

### Load data

```python
df = pd.read_csv("C:\\Users\\juhic\\OneDrive\\Desktop\\DubsTech Datathon\\data.csv")
df['collection_week'] = pd.to_datetime(df['collection_week'])
df.info()
```

### Load required columns

```python
df2 = pd.read_excel("eda.xlsx", sheet_name = "columns")
df3 = pd.read_excel("eda.xlsx", sheet_name = "base_columns")
```

```python
columns = list(df3['Name']) + list(df2['sum_columns']) + list(df2['coverage_columns'].dropna())
df = df[columns]
```

### Handle nulls & -9999 data records

```python
df.fillna(0, inplace = True)
df.replace(to_replace = -999999, value = 2, inplace = True)
```

### Create avg_columns using sum & coverage

```python
sum_cols = list(df2['sum_columns'])[0:25]
coverage_cols = list(df2['coverage_columns'])[0:25]
avg_cols = list(df2['avg_columns'].dropna())

for i in range(len(avg_cols)):
    df[avg_cols[i]] = df[sum_cols[i]] / df[coverage_cols[i]]
```

### Filter for required granularity

```python
# State filters
mask1 = df['hospital_pk'] == '052052'
mask2 = df['city'] == 'PERRIS'
mask3 = df['is_metro_micro'] == True
mask4 = df['state'] == 'CA'

# Apply filters
df_filtered = df.loc[mask4]
```
